```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json

from IPython.display import display, HTML
```

```python
df = pd.read_csv("dataset/phonecall.csv")
df
```

```python
datatest = {
    'pelanggan':'fernando',
    'kartu':'prabayar',
    'panggilan':'cukup',
    'blok':'rendah'
}
pd.DataFrame(data=[datatest])
```

```python
def get_unique(attributes):
    unique = {}
    for i, att in enumerate(attributes):
        unique[att] = set(df[att])
    return (unique)

def naive_bayes(attributes, y_title, dataset, datatest):
    unique_att = get_unique(attributes)
    unique_y = set(dataset[y_title])
    
    final_result = {}
    for y in unique_y:
        result = {}
        value = 1
        y_count = 0
        y_total = len(dataset[y_title])
        for data in dataset[y_title]:
            if data == y:
                y_count += 1
        result['probabilitas label'] = (y_count + 1)/(y_total + len(unique_y))
        value *= result['probabilitas label']
        
        for att in attributes:
            count = 0
            for i in range(len(dataset)):
                x_train = dataset[att][i]
                y_train = dataset[y_title][i]
                if x_train == datatest[att] and y_train == y:
                    count += 1
            result[att] = (count + 1)/(y_count + len(unique_att[att]))
            value *= result[att]
            
        result['probabilitas akhir'] = value
        final_result[y] = result
                
    return final_result
```

```python
attributes = ['kartu','panggilan', 'blok']
results = naive_bayes(attributes, 'bonus', df, datatest)
pd.DataFrame(data=results)
```
